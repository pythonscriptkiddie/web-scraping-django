```python
from typing import List, Tuple, Dict
from dask.distributed import Client
from dask.diagnostics import ProgressBar
import requests
from bs4 import BeautifulSoup
```

```python
client = Client(asynchronous=True)
```

```python
raw_data = requests.get('http://www.bobbybandz.com/')
```

```python
soup = BeautifulSoup(raw_data.text, 'lxml')
```

```python
title = soup.find('title')
```

```python
links = soup.find_all({'a':'href'})
```

```python
page_urls = [
            ('home', 'http://www.bobbybandz.com'),
            ('weddings', 'http://www.bobbybandz.com/weddings'),
            ('videos','http://www.bobbybandz.com/videos')
            ]
```

```python
def scrape_page(page_meta: Tuple) -> Dict:
    page_name, url = page_meta
    result = requests.get(url)
    return {
        'page_name': page_name,
        'url': url,
        'response': result
            }
```

```python
scraped_urls = client.map(scrape_page, page_urls)
```

```python
scraped_urls2 = await client.gather(scraped_urls)
```

```python
def make_soup(page_meta: Dict) -> Dict:
    soup = BeautifulSoup(page_meta['response'].text, 'lxml')
    page_meta['response'] = str(BeautifulSoup(page_meta['response'].text, 'lxml'))
    return page_meta
```

```python
scraped_urls3 = client.map(make_soup, scraped_urls2)
```

```python
scraped_urls4 = await client.gather(scraped_urls3)
```
